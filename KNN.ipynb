```python
import math
import matplotlib.pyplot as plt
import numpy as np
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, ConfusionMatrixDisplay
from sklearn.model_selection import train_test_split
from sklearn.datasets import make_blobs, load_iris, load_wine
from sklearn.neighbors import KNeighborsClassifier
from collections import Counter

random_seed=5

def gerar_amostras():
    #centro de massa da distribuio
    centers = [[2, 2], [2, 6], [6, 5]]
    n_classes = len(centers)

    #criao da distribuio (300 instncias 2 features)
    data, labels = make_blobs(n_samples=300, n_features=2,
                            centers=np.array(centers),
                            random_state=random_seed)

    amostras, amostras_teste, classes, classes_teste = train_test_split(data, labels,
                        train_size=0.7,
                        test_size=0.3,
                        random_state=random_seed)

    #Exibio
    colours = ('green', 'red', 'blue')
    legendas = ('Verde', 'Vermelho', 'Azul')
    n_classes = 3

    fig, ax = plt.subplots()
    for n_class in range(0, n_classes):
        ax.scatter(amostras[classes==n_class, 0], amostras[classes==n_class, 1],
                c=colours[n_class], s=10, label=legendas[n_class])

    ax.legend(loc='upper right')
    plt.show()
    return amostras, amostras_teste, classes, classes_teste

class KNNModel:
    def fit(self, amostras, classes):
        self.amostras = amostras
        self.classes = classes

    def distancia_euclidiana(self, amostra1, amostra2):
        soma = 0
        for i in range(len(amostra1)):
            diferenca = amostra1[i] - amostra2[i]
            soma += diferenca ** 2

        distancia = math.sqrt(soma)
        return distancia

    def predict(self, amostra, k):
        distancias = []
        for i in range(len(self.amostras)):
            distancia = self.distancia_euclidiana(amostra, self.amostras[i])
            distancias.append((round(distancia, 2), self.classes[i]))
        distancias.sort(key=lambda x: x[0])

        # Separar tupla em array com apenas classes
        classes = [classe[1] for classe in distancias][:k]

        moda = Counter(classes).most_common()[0][0]
        return moda

def teste_random(k):
    amostras, amostras_teste, classes, classes_teste = gerar_amostras()
    knn = KNNModel()
    knn.fit(amostras, classes)

    classes_geradas = []
    for i in range(len(amostras_teste)):
        amostra = amostras_teste[i]
        classe = classes_teste[i]
        resultado = knn.predict(amostra, k)
        ##print(f"Ponto {ponto} , predict {resultado}, real {classe}", "OK" if classe == resultado else "FAIL")
        classes_geradas.append(resultado)

    print("Accuracy:", accuracy_score(classes_teste, classes_geradas))
    print(classification_report(classes_teste, classes_geradas))

    labels = (0, 1, 2)
    cm = confusion_matrix(classes_teste, classes_geradas, labels=labels)
    disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=('Verde', 'Vermelho', 'Azul'))
    disp.plot(cmap=plt.cm.Blues)
    plt.show()

    knn_scikit = KNeighborsClassifier(metric='minkowski',
                                      p=2,
                                      n_neighbors=k)


    knn_scikit.fit(amostras, classes)

    classes_geradas = knn_scikit.predict(amostras_teste)

    print("Accuracy scikit:", accuracy_score(classes_teste, classes_geradas))
    print(classification_report(classes_teste, classes_geradas))

def teste_vinho(k):
    data, labels = load_wine(return_X_y=True)
    amostras, amostras_teste, classes, classes_teste = train_test_split(data, labels,
                        train_size=0.7,
                        test_size=0.3,
                        random_state=random_seed)

    #Exibio
    colours = ('green', 'red', 'blue')
    legendas = ('0', '1', '2')
    n_classes = 3
    fig, ax = plt.subplots()
    for n_class in range(0, n_classes):
        ax.scatter(amostras[classes==n_class, 0], amostras[classes==n_class, 1],
                c=colours[n_class], s=10, label=legendas[n_class])
    ax.legend(loc='upper right')
    plt.show()

    knn = KNNModel()
    knn.fit(amostras, classes)

    classes_geradas = []
    for i in range(len(amostras_teste)):
        amostra = amostras_teste[i]
        classe = classes_teste[i]
        resultado = knn.predict(amostra, k)
        ##print(f"Ponto {ponto} , predict {resultado}, real {classe}", "OK" if classe == resultado else "FAIL")
        classes_geradas.append(resultado)

    print("Accuracy:", accuracy_score(classes_teste, classes_geradas))
    print(classification_report(classes_teste, classes_geradas))

    labels = (0, 1, 2)
    cm = confusion_matrix(classes_teste, classes_geradas, labels=labels)
    disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=legendas)
    disp.plot(cmap=plt.cm.Blues)
    plt.show()

    # ----------------
    #p=1 ~ Manhattan Distance
    #p=2 ~ Euclidean
    #p!=1|2 ~ minkowski
    knn_scikit = KNeighborsClassifier(metric='minkowski',
                                      p=2,
                                      n_neighbors=k)


    knn_scikit.fit(amostras, classes)

    classes_geradas = knn_scikit.predict(amostras_teste)

    print("Accuracy scikit:", accuracy_score(classes_teste, classes_geradas))
    print(classification_report(classes_teste, classes_geradas))


def teste_iris(k):
    data, labels = load_iris(return_X_y=True)
    amostras, amostras_teste, classes, classes_teste = train_test_split(data, labels,
                        train_size=0.7,
                        test_size=0.3,
                        random_state=random_seed)

    #Exibio
    colours = ('green', 'red', 'blue')
    legendas = ('setosa', 'versicolor', 'virginica')
    n_classes = 3
    fig, ax = plt.subplots()
    for n_class in range(0, n_classes):
        ax.scatter(amostras[classes==n_class, 0], amostras[classes==n_class, 1],
                c=colours[n_class], s=10, label=legendas[n_class])
    ax.legend(loc='upper right')
    plt.show()

    knn = KNNModel()
    knn.fit(amostras, classes)

    classes_geradas = []
    for i in range(len(amostras_teste)):
        amostra = amostras_teste[i]
        classe = classes_teste[i]
        resultado = knn.predict(amostra, k)
        ##print(f"Ponto {ponto} , predict {resultado}, real {classe}", "OK" if classe == resultado else "FAIL")
        classes_geradas.append(resultado)

    print("Accuracy:", accuracy_score(classes_teste, classes_geradas))
    print(classification_report(classes_teste, classes_geradas))

    labels = (0, 1, 2)
    cm = confusion_matrix(classes_teste, classes_geradas, labels=labels)
    disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=legendas)
    disp.plot(cmap=plt.cm.Blues)
    plt.show()

    # ----------------
    #p=1 ~ Manhattan Distance
    #p=2 ~ Euclidean
    #p!=1|2 ~ minkowski
    knn_scikit = KNeighborsClassifier(metric='minkowski',
                                      p=2,
                                      n_neighbors=k)


    knn_scikit.fit(amostras, classes)

    classes_geradas = knn_scikit.predict(amostras_teste)

    print("Accuracy scikit:", accuracy_score(classes_teste, classes_geradas))
    print(classification_report(classes_teste, classes_geradas))

print("Teste Random")
print("=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=")
print("k=1")
teste_random(1)
print("=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=")
print("k=5")
teste_random(5)
print("=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=")
print("k=7")
teste_random(7)
print("=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=")
print("Teste Vinho")
print("=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=")
print("k=1")
teste_vinho(1)
print("=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=")
print("k=5")
teste_vinho(5)
print("=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=")
print("k=7")
teste_vinho(7)
print("=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=")
print("Teste Iris")
print("=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=")
print("k=1")
teste_iris(1)
print("=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=")
print("k=5")
teste_iris(5)
print("=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=")
print("k=7")
teste_iris(7)
print("=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=")
```
